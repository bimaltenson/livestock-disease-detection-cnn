# Methodology

The proposed methodology consists of five main stages: data preprocessing, data augmentation, model development, model training, and model evaluation. Two deep learning models, a custom Convolutional Neural Network (CNN) and MobileNetV2, were implemented and compared for cow disease classification.

## 1. Data Preprocessing

Before training the models, the input images were preprocessed to ensure consistency and improve model performance.

- Images were resized to **224 × 224 pixels**.
- Pixel values were normalized using **rescale = 1/255**.
- The dataset was divided into **training** and **validation** sets.

## 2. Data Augmentation

To improve model generalization and reduce overfitting, data augmentation techniques were applied during training.

The following augmentation methods were used:

- Rotation Range = 20°
- Zoom Range = 0.2
- Horizontal Flip = True

These techniques generate different variations of the training images while preserving their class labels.

## 3. CNN Model Development

A custom Convolutional Neural Network (CNN) was designed for multiclass image classification. The network consists of:

- Conv2D Layer (32 filters)
- MaxPooling2D Layer
- Conv2D Layer (64 filters)
- MaxPooling2D Layer
- Conv2D Layer (128 filters)
- MaxPooling2D Layer
- Flatten Layer
- Dense Layer (128 neurons, ReLU activation)
- Dropout Layer (0.5)
- Output Layer (3 neurons with Softmax activation)

The CNN automatically learns important visual features from cow disease images.

## 4. MobileNetV2 Model Development

A transfer learning approach was implemented using the pretrained MobileNetV2 architecture.

The MobileNetV2 model consists of:

- Pretrained MobileNetV2 (ImageNet weights)
- GlobalAveragePooling2D Layer
- Dense Layer (128 neurons, ReLU activation)
- Dropout Layer (0.5)
- Output Layer (3 neurons with Softmax activation)

The pretrained model utilizes knowledge learned from ImageNet to improve feature extraction for cow disease classification.

## 5. Model Training and Evaluation

Both models were trained using the training dataset and evaluated using the validation dataset.

The performance of each model was measured using:

- Accuracy
- Precision
- Recall
- F1-score
- Confusion Matrix

Finally, the performance of the custom CNN and MobileNetV2 models was compared to determine the most effective approach for cow disease classification.

```python
import os

dataset_path = "/kaggle/input/datasets/bimaltenson/cows-disease-dataset/Cows datasets"

classes = os.listdir(dataset_path)

print("Classes:", classes)

for cls in classes:
    folder = os.path.join(dataset_path, cls)
    print(f"{cls}: {len(os.listdir(folder))} images")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

dataset_path = "/kaggle/input/datasets/bimaltenson/cows-disease-dataset/Cows datasets"
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

train_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

print(train_data.class_indices)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense, Dropout

cnn_model = Sequential([

    Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.5),

    Dense(3, activation='softmax')

])

cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

cnn_model.summary()
```

### Figure 1. CNN Model Architecture

Figure 1 illustrates the architecture of the proposed custom Convolutional Neural Network (CNN) used for cow disease classification. The model consists of three convolutional layers with max-pooling operations for feature extraction, followed by a flatten layer, a dense layer with ReLU activation, a dropout layer for regularization, and a Softmax output layer for classifying images into three disease categories.

```python
history = cnn_model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
import matplotlib.pyplot as plt
import os

os.makedirs("/kaggle/working/outputs", exist_ok=True)

# Accuracy curve
plt.figure(figsize=(8, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('CNN Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('/kaggle/working/outputs/cnn_accuracy_curve.png')
plt.show()

# Loss curve
plt.figure(figsize=(8, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('CNN Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.savefig('/kaggle/working/outputs/cnn_loss_curve.png')
plt.show()
```

### Figure 2. CNN Training and Validation Accuracy and Loss Curves

Figure 2 presents the training and validation accuracy and loss curves of the custom CNN model over 10 training epochs. The training accuracy increased steadily while the training loss consistently decreased, indicating that the model effectively learned meaningful features from the training data. The validation accuracy also improved throughout training, while the validation loss showed only minor fluctuations, suggesting that the model generalized well to unseen validation data with limited overfitting.

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Reset validation generator
val_data.reset()

# Predictions
predictions = cnn_model.predict(val_data)

y_pred = np.argmax(predictions, axis=1)
y_true = val_data.classes

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=list(val_data.class_indices.keys()),
    yticklabels=list(val_data.class_indices.keys())
)

plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print(classification_report(
    y_true,
    y_pred,
    target_names=list(val_data.class_indices.keys())
))
```
